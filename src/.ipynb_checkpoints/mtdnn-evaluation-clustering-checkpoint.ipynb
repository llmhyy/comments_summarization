```python
def get_score_tuples(score_list):
    '''
    :param score-list: array-like of probabilities of each class, like the mt-dnn outputs.
    :returns: a list of tuples. each tuple contains the probabilities of each class for a single prediction.
    '''
    tuples = []
    for i in range(0, len(score_list), 3):
        tuples.append((score_list[i], score_list[i+1], score_list[i+2]))
    return tuples
```

```python
def compute_distances(tuples):
    '''
    :param tuples: a list of 3-ples. [0] is contradiction, [1] is neutral, [2] is entailment.
    :returns: a list of distances.
    '''
    distances = []
    for result in tuples:
        C = result[0]
        N = result[1]
        E = result[2]
        if C > 0.5:
            distances.append(C)
        elif C == E:
            distances.append(0.5)
        else:
            distances.append(N/(N+E)*0.5)
    return distances
```

```python
tuples = get_score_tuples(results2['scores'])
distances = compute_distances(tuples)
```

```python
import numpy as np
import pandas as pd

import scipy
from scipy.cluster.hierarchy import dendrogram,linkage
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import sklearn
# from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
import sklearn.metrics as sm
from sklearn.preprocessing import scale
```

```python
def cluster(df_clean):
    # obtain unique sentences from pairwise labelled dataset
    first = np.unique(df_clean['first'].values)
    second = np.unique(df_clean['second'].values)
    combined_arr = np.concatenate((first, second))
    sentences = np.unique(combined_arr)
    print(len(sentences))

    # df where sentence i is paired with sentence i and their relative dist=0.
    add_data = np.column_stack([sentences, sentences, np.zeros(len(sentences))])
    add_df = pd.DataFrame(data=add_data, columns=df_clean.columns)
    final_df = df_clean.append(add_df, ignore_index=True)

    # pairwise dist matrix
    piv_df = final_df.pivot(col1_name, col2_name, col3_name).fillna(0)
    piv_arr = piv_df.values
    dist_mat = piv_arr + np.transpose(piv_arr)
    condensed_mat = squareform(dist_mat)

    linkage_arr = linkage(condensed_mat)
    # dendrogram(linkage_arr,
    #             orientation='top',
    #             labels=piv_df.index.values,
    #             distance_sort='descending',
    #             show_leaf_counts=True)

    from sklearn.metrics import silhouette_score
    sil_scores = []
    for i in range(2, len(dist_mat)):
        clusters = AgglomerativeClustering(n_clusters=i, affinity='precomputed', linkage='single')
        clusters.fit_predict(dist_mat)

        silhouette_avg = silhouette_score(dist_mat, clusters.labels_, metric='precomputed')
        sil_scores.append(silhouette_avg)
        # print(i, ": ", silhouette_avg)

    sil_scores_arr = np.asarray(sil_scores)
    print('max avg silhouette score: ', sil_scores_arr.max())
    optimal_k = np.where(sil_scores_arr == sil_scores_arr.max())[0][0] + 2
    print('nclusters w max avg silhouette score: ', optimal_k)
    
    return optimal_k, piv_df
```

```python
col1_name = 'first' 
col2_name = 'second'
col3_name = 'predicted_distances'
df_clean = truth_df[[col1_name, col2_name, col3_name]]

optimal_k, piv_df = cluster(df_clean)

# observe 'optimal' clustering
clusters = AgglomerativeClustering(n_clusters=optimal_k, affinity='precomputed', linkage='single')
clusters.fit_predict(dist_mat)
for i in range(clusters.n_clusters_):
    indices = np.where(clusters.labels_ == i)[0]
    
    print(f'cluster {i} (size {len(indices)}):')
    print(indices)
    for idx in indices:
        display(piv_df.index[idx])
    print('\n')
```
