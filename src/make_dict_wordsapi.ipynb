```python
import json
import numpy as np
import networkx as nx
```

```python
with open("../data/wordsapi_yihao_output.json") as f:
    crawler_output = json.load(f)
len(crawler_output)
```

```python
trait_defs = dict()
for output in crawler_output:
    trait = output['word']
    defs = list()
    for result in output['definitions']:
        if result['partOfSpeech'] == 'adjective':
            defs.append(result['definition'])
    trait_defs[trait] = defs
trait_defs
```

```python
path = input('enter path to save dictionary in json format:\n')
with open(path, 'w') as json_file:
  json.dump(trait_defs, json_file)
```

```python
count = 0
for defs in trait_defs:
    count = count + len(defs)
print(count)
```
