```python
from nltk import sent_tokenize
```

```python
# clean text.
all_sents = list()
for i in range(len(df_tmp)):
    text = df_tmp.iloc[i]['Text']
    text = text.replace("<br />", "")
    sents = sent_tokenize(text)
    all_sents.extend(sents)
print(all_sents)
```

```python
len(all_sents)
```

```python
to_save = pd.DataFrame(data=all_sents, columns=["sentence"])
to_save.head()
```

```python
path = input("enter path:\n")
to_save.to_csv(path)
```
