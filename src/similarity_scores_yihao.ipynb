```python
MODEL_NAME = "bert-base-cased"

# We need to create the model and tokenizer
model = AutoModel.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

pooled_list = []
for node1, node2, edge in G.edges(data='relation'):
    if edge == 'defined as':
        tokens = tokenizer.tokenize(node2)
        tokens_ids = tokenizer.convert_tokens_to_ids(tokens)

        tokens_ids = tokenizer.build_inputs_with_special_tokens(tokens_ids)


        tokens_pt = torch.tensor([tokens_ids])
        _, pooled_temp = model(tokens_pt)
        pooled_temp = pooled_temp.squeeze(0)
        pooled_list.append((node1, node2, pooled_temp))
```

```python
file_postfix = ['ascendant', 'jupiter', 'lilith', 'mars', 'mercury', 'moon', 'neptune', 'pluto', 'saturn', 'sun', 'uranus', 'venus']
path = '/Users/Vincent/Documents/Thesis/comments_summarization/data/planets-in-signs/'
sentences = []
for postfix in file_postfix:
    with open(path + 'planets-in-signs_{}_data.json'.format(postfix), 'r') as file:
        for line in file:
            text = json.loads(line)
            for key in text.keys():
                for descriptions in text[key]:
                    if descriptions == ' ' or 'Spiritual lesson to learn' in descriptions or 'See more' in descriptions or 'More about' in descriptions:
                        continue
                    descriptions = [s.strip() for s in descriptions.split('.') if s.strip()!='']
                    sentences += descriptions
sentences = set(sentences)
```

```python
print('\n'.join(sentences))
#print(sentences)
```

```python
def tokenize_sentence(sentence):
    tokens = tokenizer.tokenize(sentence)
    #print("Tokens: {}".format(tokens))

    # This is not sufficient for the model, as it requires integers as input, 
    # not a problem, let's convert tokens to ids.
    tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
    #print("Tokens id: {}".format(tokens_ids))

    # Add the required special tokens
    tokens_ids = tokenizer.build_inputs_with_special_tokens(tokens_ids)

    # We need to convert to a Deep Learning framework specific format, let's use PyTorch for now.
    tokens_pt = torch.tensor([tokens_ids])
    #print("Tokens PyTorch: {}".format(tokens_pt))

    # Now we're ready to go through BERT with out input
    outputs, pooled = model(tokens_pt)
    #print("Token wise output: {}, Pooled output: {}".format(outputs.shape, pooled.shape))
    pooled = pooled.squeeze(0)
    return pooled
```

```python
for s in sentences:
    pooled = tokenize_sentence(s)
    print(s)
    max = 0
    min = 1

    for node1, node2, pooled_temp in pooled_list:
        similarity = torch.cosine_similarity(pooled_temp, pooled, 0).item()
        if similarity >max:
            max = similarity
            max_node = node2
            max_word = node1
        if similarity < min:
            min = similarity
            min_node =node2
            min_word = node1
    print(max_node, max_word) 
    print(min_node, min_word)
    print()
```
