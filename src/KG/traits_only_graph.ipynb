```python
import networkx as nx
from networkx import NetworkXError
```

```python
G = nx.read_gpickle("../data/graph_cleaned.pickle")
```

```python
count = 0
for node1, node2, edge in G.edges(data='relation'):
    if edge == 'defined as':
        print(node1, '\t', node2)
        count = count + 1
print('no. of definitions: ', count)
```

```python
# might want to remove the stuff in brackets at the start of the definition, at some point.
```

```python
from copy import deepcopy
G2 = deepcopy(G)

for node1, node2, edge in G.edges(data='relation'):
    if edge == 'defined as':
        try:
            G2.remove_node(node2)
        except NetworkXError as e:
            print(e)
            pass
len(G2.nodes())
```

```python
G2.nodes()  # G2 should now be a traits-only graph.
```

```python
G2.edges.data()
```

```python
path = input("enter path to save pickled graph:\n")
nx.write_gpickle(G2, path)
```
